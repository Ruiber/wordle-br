```python
import pandas as pd
import numpy as np
```

```python
word_length = 5

base_path = '../data'
dicio_path = '/dicio'
palavras_path = '/palavras'
verbos_path = '/verbos'
conjugacoes_path = '/conjugacoes'
data_path = '/data'

def read_doc(rel_path):
    with open(base_path + rel_path, 'r') as file:
        out = {line.strip() for line in file}
    return out
```

```python
# dicio está contido em palavras!
dicio = read_doc(dicio_path)
palavras = read_doc(palavras_path)

dicio.issubset(palavras)
```

```python
# nenhum verbo de <word_length> letras não está em dicio!
verbos = read_doc(verbos_path)

[word for word in verbos.difference(dicio) if len(word) == word_length]
```

```python
# há conjugações de verbos com <word_length> letras que não estão em dicio!
conjugacoes = read_doc(conjugacoes_path)

len([word for word in conjugacoes.difference(dicio) if len(word) == word_length])
```

```python
selected_words = set(filter(lambda word: len(word) == word_length, dicio))
len(selected_words)
```

```python
word_frequency = pd.read_csv(base_path + data_path)
word_frequency
```

```python
word_frequency['tamanho'] = word_frequency.palavra.apply(len)
word_frequency
```

```python
# deixando apenas o tamanho que nos importa
word_frequency = word_frequency[word_frequency.tamanho == word_length]
word_frequency = word_frequency.drop(columns='tamanho')
word_frequency
```

```python
# todas as palavras de <word_length> letras estão com suas estatísticas realizadas em word_frequency
words = set(word_frequency.palavra.unique())

selected_words.issubset(words)
```

```python
selected_words_df = word_frequency[word_frequency.palavra.isin(selected_words)]
selected_words_df
```

```python
selected_words = sorted(list(selected_words))

with open(f'../words/palavras_{word_length}', 'w') as file:
    for word in selected_words:
        file.write(word + '\n')

selected_words_df.to_csv(f'../words/estatisticas_palavras_{word_length}', index=False)
```
